# Handle Missing Values

Filled missing values in numerical columns using the Median.

Filled missing values in categorical columns using the Mode.

```python
df1.dtypes
```

```python
# For Numarical columns
Median_dict = {
    "Age":df1["Age"].median(),
    "Attendance_Percentage":df1["Attendance_Percentage"].median(),
    "Study_Hours_per_day":df1["Study_Hours_per_day"].median(),
    "Math_Score":df1["Math_Score"].median(),
    "Science_Score":df1["Science_Score"].median(),
    "English_Score":df1["English_Score"].median(),
    "Total_Score":df1["Total_Score"].median()
}
for col in Median_dict:
  df1[col]=df1[col].fillna(Median_dict[col])
```

```python
df1.isnull().sum()
```

```python
#For Categorical columns fill by mode
cat_cols = df1.select_dtypes(include="object").columns


for col in cat_cols:
    df1[col] = df1[col].fillna(df1[col].mode()[0])
```

```python
df1.isnull().sum()
```

# Encoding
Applied Label Encoding to Gender, Internet_Access, and Result.


# 4. Train-Test Split
*Train-Test Split is a method of dividing a dataset into two parts: one part is used to train the machine learning model, and the other part is used to test how well the model works on new data.*


# Comment
We used an 80/20 split, where 80% of the data is used for training the model and 20% is reserved for testing. This ratio provides enough data for learning while keeping sufficient unseen data to evaluate the model.

```python
#For linear regression
from sklearn.model_selection import train_test_split
X=df1.drop("Total_Score",axis=1)
y=df1["Total_Score"]
X_train,X_test,y_train,y_test=train_test_split(
  X,
  y,
  test_size=0.20,
  random_state=42)
print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.model_selection import train_test_split
X = df1.drop(["Result"], axis=1)
y = df1["Result"]
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)
```

After using stratify=y, the class distribution in the training and testing datasets remains nearly the same as the original dataset. This confirms that stratification was applied successfully.

```python
print(y.value_counts())
print(y_train.value_counts())
print(y_test.value_counts())
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt

# Predicted Probability
y_prob = longistic_model.predict_proba(X_test)[:,1]

# ROC Curve
fpr, tpr, thresholds = roc_curve(y_test, y_prob)

# AUC Score
auc = roc_auc_score(y_test, y_prob)
print("AUC Score:", auc)

# Plot ROC Curve
plt.figure(figsize=(6,5))
plt.plot(fpr, tpr, label=f"AUC = {auc:.3f}")
plt.plot([0, 1], [0, 1], "r--")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.show()
```
